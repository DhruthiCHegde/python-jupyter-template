### Exercise 4: Pie Chart with Percentages

```python
# TASK: Create a pie chart with the following data: labels = ['Python', 'Java', 'C++', 'JavaScript'] and sizes = [40, 25, 20, 15].
# Display the percentages on the chart using autopct.
import matplotlib.pyplot as plt
import numpy as np

x = np.arange(0, 11)

y = x ** 2

plt.plot(x, y, marker='o', linestyle='-', color='b', label='y = x^2')

plt.title('Square of X')
plt.xlabel('X values')
plt.ylabel('Y values (X squared)')
plt.grid(True)
plt.legend()


plt.show()

categories = ['A', 'B', 'C', 'D']
values = [5, 7, 3, 9]
colors = ['red', 'green', 'blue', 'purple']

plt.bar(categories, values, color=colors)

plt.title('Category Values')
plt.xlabel('Categories')
plt.ylabel('Values')

plt.show()

x_scatter = [1, 2, 3, 4, 5]
y_scatter = [2, 4, 6, 8, 10]

plt.scatter(x_scatter, y_scatter, color='blue')


for i, txt in enumerate(zip(x_scatter, y_scatter)):
    plt.annotate(f'{txt}', (x_scatter[i], y_scatter[i]), textcoords="offset points", xytext=(5,5), ha='right')

plt.title('Scatter Plot with Annotations')
plt.xlabel('X values')
plt.ylabel('Y values')
plt.grid(True)

plt.show()

labels = ['Python', 'Java', 'C++', 'JavaScript']
sizes = [40, 25, 20, 15]


plt.pie(sizes, labels=labels, autopct='%1.1f%%', colors=['blue', 'green', 'red', 'purple'], startangle=140)

plt.title('Programming Language Popularity')


plt.show()
```
